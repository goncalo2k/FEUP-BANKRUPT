### Filtering Outliers
By making the previous checks, we know that:
 - The ID column is quite useless for our study, and therefore should be excluded;
 - There are rows where there is no age, job or balance value - these rows should be pruned;
 - Besides the previously pruned, we should also prune two other columnds: the "poutcome" column and the "contact" column, since both of them have "unknown" as all their values;

```python
df = df.drop(["Id", "poutcome", "contact"], axis="columns")
```
